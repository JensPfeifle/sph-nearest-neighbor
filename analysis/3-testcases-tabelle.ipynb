```python
import pandas as pd
```

## Read in dataframe

```python
df = pd.read_csv("results_df.csv")
df.head()
```

## Test Case Tabelle

```python
def test(array, value):
    if value in array:
        return r"$\bullet$"
    return " "

testann = lambda n: test(n, "ANN")
testannnl = lambda n: test(n, "ANN-NL")
testcll = lambda n: test(n, "CLL")


latexdf = df.groupby(['filltype','fill'])["method"].unique().to_frame()
latexdf["ANN"] = latexdf["method"].apply(testann)
latexdf["ANN-NL"] = latexdf["method"].apply(testannnl)
latexdf["CLL"] = latexdf["method"].apply(testcll)
latexdf = latexdf.drop(columns = ["method"])
latexdf
```

## Latex

```python
outputfile = "../latex/kapitel/testcasestabelle.tex"
floatfmt = "%.2f" 
groups_spacing = "2.0ex"


latexsrc = latexdf.to_latex(float_format=floatfmt, escape=False).splitlines()

for n in range(6,len(latexsrc)-3):
    if not latexsrc[n][0] == " ":
        latexsrc[n-1] = latexsrc[n-1] + "[{}]".format(groups_spacing)

with open(outputfile,'w') as f:
    for line in latexsrc:
        f.write(line + '\n')
```
