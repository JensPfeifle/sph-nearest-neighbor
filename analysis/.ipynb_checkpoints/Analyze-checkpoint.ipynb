## Read data

```python
df = pd.read_csv('stats.csv')
EDITED = False
```

```python
if EDITED:
    print("Not reapplying cleanup.")
else:
    EDITED = True
    # fill to percent
    df['fill'] = (df['fill']*100).astype(int)
    df['fill']
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax = sns.stripplot(x="filltype", y="ttotal", data=df, ax=ax)
```
