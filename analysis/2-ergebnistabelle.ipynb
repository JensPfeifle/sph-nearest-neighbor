```python
import pandas as pd
```

## Read in dataframe

```python
df = pd.read_csv("results_df.csv")
df.head()
```

## Ergebnistabelle
Grouped mean values

```python
latexdf = df.groupby(["fill", "filltype", "method"]).mean()
latexdf = latexdf.drop(columns = ["listmethod"])
latexdf = latexdf.reindex(['ndatapts', 'memory',
       'tfrsearch', 'tksearch', 'tprocessing', 'ttotal'], axis=1)
latexdf = latexdf.rename(index=str, columns={'ndatapts':"Datapts",
                              'memory':"Memory",
                              'tfrsearch':"$t_{frsearch}$",
                              'tksearch':"$t_{tksearch}$",
                              'tprocessing':"$t_{processing}$",
                              'ttotal':"$t_{total}$"})

latexdf
```

## Latex

```python
outputfile = "../latex/kapitel/ergebnistabelle.tex"
floatfmt = "%.2f" 
groups_spacing = "2.0ex"


latexsrc = latexdf.to_latex(float_format=floatfmt, escape=False).splitlines()

for n in range(6,len(latexsrc)-3):
    if not latexsrc[n][0] == " ":
        latexsrc[n-1] = latexsrc[n-1] + "[{}]".format(groups_spacing)

with open(outputfile,'w') as f:
    for line in latexsrc:
        f.write(line + '\n')
```
